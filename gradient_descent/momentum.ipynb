# Momentum Gradient Descent
Import necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
def f(x):
    return x ** 2 + 10 * math.sin(x)
```

```python
def f_deriv(x):
    return 2*x + 10*math.cos(x)
```

```python
def gradient_descent(f, init_x, learning_rate):
    count = 0
    x_history = [init_x]
    for i in range(1000):
        count += 1
        x_new = x_history[-1] - learning_rate * f_deriv(x_history[-1])
        if abs(f_deriv(x_new)) < 1e-3:
            break
        x_history.append(x_new)
    return x_history[-1], x_history, count
```

```python
sol, sol_history, iters_count = gradient_descent(f, -5,0.1)
sol
```

```python
x1 = np.arange(-5, 6, 0.1)
y1 = np.array(list(map(f, x1)))
plt.figure(figsize=(20,10))
plt.plot(x1, y1)
plt.scatter(sol, f(sol), c='r', s=50)
plt.show()
```
