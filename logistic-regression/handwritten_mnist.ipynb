```python
from sklearn.datasets import load_digits
```

```python
digits = load_digits()
```

```python
# 1797 images(8x8=64)
print('Image data: ', digits.data.shape)
print('Label (0-9)', digits.target.shape)
```

# Showing the images and the labels(digits dataset)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(20, 4))
for index, (img_data, label) in enumerate(zip(digits.data[0:5], digits.target[0:5])):
    plt.subplot(1, 5, index+1)
    plt.imshow(np.reshape(img_data, (8, 8)), cmap=plt.cm.gray)
    plt.title('Training: {}'.format(label))
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, x_test, y_train, y_test = train_test_split(digits.data, digits.target, test_size=0.25, random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf = LogisticRegression()
clf.fit(X_train, y_train)
```

```python
clf.predict(x_test[0:9])
```

```python
plt.figure()
plt.imshow(x_test[0].reshape((8, 8)), cmap=plt.cm.gray)
```

```python
score = clf.score(x_test, y_test)
print(score)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import metrics
```

```python
predictions = clf.predict(x_test)
```

```python
cm = metrics.confusion_matrix(y_test, predictions)
print(cm)
```
